```python
import graphlab
```

```python
p = graphlab.SFrame.read_csv('prototipo.txt',sep='\t',header=False)
```

```python
import re
def clean(s):
     s = re.sub('\[u\]','',s)
     s = re.sub('\[.\]','',s)
     s = re.sub('\[','',s)
     s = re.sub('\]','',s)
     s = re.sub('\*',' ',s)
     s = re.sub('\-',' ',s)
     s = re.sub('  *',' ',s)
     s = re.sub('^ ','',s)
     s = re.sub(' $','',s)
     s = re.sub('[()]',' ',s)
     return s
```

```python
p.head()
```

```python
p.rename({'X1':'id','X2':'text','X3': 'meaning'})
```

```python
p.head()
```

```python
p['clean'] = p['text'].apply(clean)
```

```python
p.head()
```

```python
p['words'] = graphlab.text_analytics.count_words(p['clean'])
```

```python
p.head()
```

```python
model = graphlab.topic_model.create(p['words'],num_topics=200,num_iterations=300)
```

```python
p['topic'] = model.predict(p['words'])
```
